```python
import os
import glob
import pandas as pd
import chardet

folder_path = './'
csv_files = glob.glob(os.path.join(folder_path, '경찰청_음주운전적발기록_*.csv'))

dataframes = []

for file in csv_files:
    try:
        # 인코딩 감지
        with open(file, 'rb') as f:
            result = chardet.detect(f.read(10000))
            encoding = result['encoding']
            print(f"[INFO] {file} 인코딩 감지됨: {encoding}")

        # 연도 추출
        filename = os.path.basename(file)
        year = int(filename.split('_')[-1].split('.')[0])

        # 탭 구분된 파일은 sep 지정 필요
        if year in [2020, 2021]:
            df = pd.read_csv(file, encoding=encoding, sep='\t')
        else:
            df = pd.read_csv(file, encoding=encoding)

        print(f"[INFO] {filename} 컬럼명: {df.columns.tolist()}")
        dataframes.append((year, df))

    except Exception as e:
        print(f"[ERROR] {file} 읽기 실패: {e}")
```

```python
# 통합할 기준 컬럼
common_columns = ['성별', '적발횟수', '나이', '알콜농도', '측정거부여부', '측정일시', '관할경찰서']

# 컬럼명 매핑 (다른 이름들 통일)
column_mapping = {
    '측정거부': '측정거부여부',
    '연번': None,  # 제거할 컬럼
}

# 정리된 데이터프레임 리스트
cleaned_dataframes = []

for year, df in dataframes:
    # 컬럼명 매핑 및 불필요한 컬럼 제거
    df = df.rename(columns=column_mapping)
    df = df[[col for col in df.columns if col in common_columns]]

    # 누락된 컬럼 채우기 (없는 컬럼은 NaN으로 추가)
    for col in common_columns:
        if col not in df.columns:
            df[col] = pd.NA

    # 컬럼 순서 통일
    df = df[common_columns]
    df['연도'] = year  # 분석을 위한 연도 컬럼 추가
    cleaned_dataframes.append(df)
```

```python
merged_df = pd.concat(cleaned_dataframes, ignore_index=True)
merged_df.to_csv('cleaned_merged_output.csv', index=False, encoding='utf-8-sig')
print("✅ 병합 완료 → cleaned_merged_output.csv")
```

```python
df2['나이'] = pd.to_numeric(df2['나이'], errors='coerce')
old_age_df = df2[df2['나이'] >= 100]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# 한글 폰트 설정 (윈도우)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 마이너스 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(8, 4))
sns.boxplot(x=df2['나이'])

plt.title('나이 분포 박스플롯')
plt.xlabel('나이')
plt.show()
```

```python
# NaN 값 제거한 후 사분위수 계산
q1 = df2['나이'].dropna().quantile(0.25)
q3 = df2['나이'].dropna().quantile(0.75)

print(f"1사분위수 (Q1): {q1}")
print(f"3사분위수 (Q3): {q3}")
# IQR (Interquartile Range: Q3 - Q1)
iqr = q3 - q1
print(f"IQR (Q3 - Q1): {iqr}")
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

#90살까지 운전할수 있다고 생각되어 수정
outliers = df2[(df2['나이'] < lower_bound) | (df2['나이'] > 90)]

print(lower_bound)
print(upper_bound)

print(outliers)
```

```python
#연도별로고르게 분포되어있고 죽을정도는 아니라(0.4부터 죽을수도..) 제거할 기준이 명확하지 않아 일단 그대로 진행 

# NaN 값 제거한 후 사분위수 계산
a_q1 = df2['알콜농도'].dropna().quantile(0.25)
a_q3 = df2['알콜농도'].dropna().quantile(0.75)

print(f"1사분위수 (Q1): {a_q1}")
print(f"3사분위수 (Q3): {a_q3}")
# IQR (Interquartile Range: Q3 - Q1)
a_iqr = a_q3 - a_q1
print(f"IQR (Q3 - Q1): {a_iqr}")
a_lower_bound = a_q1 - 1.5 * a_iqr
a_upper_bound = a_q3 + 1.5 * a_iqr
a_outliers = df2[(df2['알콜농도'] < a_lower_bound) | (df2['알콜농도'] > a_upper_bound)]
print(f"lowwer : {a_lower_bound}")
print(f"upper : {a_upper_bound}")
print(a_outliers)
```
